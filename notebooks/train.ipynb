```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("../data/pizzas-diameter.csv")
```

```python
plt.scatter(df["diametro"], df["preco"])

# Adicionar rótulos aos eixos
plt.xlabel("Diâmetro da Pizza (cm)")
plt.ylabel("Preço (R$)")
plt.title("Relação entre Diâmetro e Preço da Pizza")

# Mostrar o gráfico
plt.show()
```

```python
modelo = linear_model.LinearRegression()

# Preparar os dados para treinamento

X = df[["diametro"]]
y = df["preco"]

# Treinar o modelo

modelo.fit(X, y)

print(modelo.predict([[110]])[0])
```
